```python
import pickle
import numpy as np

file_path = '/media/zzs/4TB/qingliu/qing_intermediate/feat_pickle/'
magic_thh = 0.7

layer_feature_dist = []
sub_type = []
view_point = []

fname = file_path + 'res_info_car_train.pickle'
with open(fname, 'rb') as fh:
    l, s, v = pickle.load(fh)
    layer_feature_dist += l
    sub_type += s
    view_point += v
    
N = len(sub_type)
print('total number of instances {0}'.format(N))

layer_feature_b = [None for nn in range(N)]
for nn in range(N):
    layer_feature_b[nn] = (layer_feature_dist[nn]<magic_thh).astype(int)
    
```

```python
layer_feature_ba = [None for nn in range(N)]
for nn in range(N):
    layer_feature_ba[nn] = np.mean(layer_feature_b[nn], axis=(1,2))
    
layer_feature_ba = np.array(layer_feature_ba)
rst_mean = np.mean(layer_feature_ba,axis=0)
best_vc = np.argsort(rst_mean)[-30:]
best_vc
```

```python
view_point = np.array(view_point)
index_set = [None for ss in range(6)]
ff_set = [None for ss in range(6)]
for ss in range(6):
    index_set[ss] = np.where(np.logical_and(view_point>ss*60, view_point<ss*60+60))[0]
    ff_set[ss] = [layer_feature_b[nn] for nn in index_set[ss]]
    
```

```python
ff_set_padded = [[] for ss in range(6)]
for ss in range(6):
    size_all = np.array([fpatch.shape[1:3] for fpatch in ff_set[ss]])
    size_all = np.max(size_all,axis=0)
    heat_map = np.zeros(size_all)
    for fpatch in ff_set[ss]:
        hi, wi = fpatch.shape[1:3]
        h_diff = size_all[0]-hi
        h_upper = int(h_diff/2)
        h_low = int(h_diff - h_upper)
        
        w_diff = size_all[1]-wi
        w_upper = int(w_diff/2)
        w_low = int(w_diff - w_upper)
        
        padded = np.pad(fpatch, ((0,0),(h_upper, h_low),(w_upper, w_low)), 'constant', constant_values=0)
        ff_set_padded[ss].append(padded)
        
    ff_set_padded[ss]=np.array(ff_set_padded[ss])
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

f, axes = plt.subplots(7,6, sharex=True, sharey=True, figsize=(25,10))
f.tight_layout() 
best_vc = np.array([2,16,39,61, 173, 108, 165])
# 2: wheel 
# 16:bumper and frame
# 39: frame,fender?
# 61: flat metal
# 173: light
# 108: car head
# 165: car head

for ss in range(6):
    heat_map = np.sum(ff_set_padded[ss],axis=0)/ff_set_padded[ss].shape[0]
    heat_map = heat_map[best_vc]
    for nn in range(heat_map.shape[0]):
        axes[nn][ss].imshow(heat_map[nn].T, interpolation='none')
        axes[nn][ss].axis('off')
        
plt.show()

        
```
